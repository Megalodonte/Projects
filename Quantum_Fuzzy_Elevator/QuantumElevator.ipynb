## <font color="green"> Problem Discussion </font>

The following notebook illustrates the capabilities of a Fuzzy Inference Engine powered by quantum computation and compares its performances to those of a classical engine.

The experiment consists in:
- Control of a virtual elevator through a simulated environment
- Elevator controlled using a **Mamdani MISO** (**M**ultiple **I**nputs **S**ingle **O**utput) fuzzy inference engine
- Two input variables are the *distance* of the elevator from the target height and its *velocity*
- Output of the engine is the *driving* (updirected) *force* that simulates a torque on the cabin pulling it upwards

(Note that the applied force can only be directed upwards, so, in order to move the elevator down, the gravitational force will be the only driver, also, the gravitational constant $g$ is set to $10m/s^2$ for simplicity and symmetry motivations)

<center><img src="Images/elevator.png" alt="Drawing" width="500" />Figure 1: Elevator diagram </center>

- Time of execution is discretized, at each iteration ($ {\delta t} = 0.01s $ simulated time) the position and velocity are updated depending on the previous state and the engine output, following the equation:

$$
x_{t+1} = x_t + v_t \delta t - {1 \over 2} {F_g \over m} {\delta t}^2 + {1 \over 2} {F_{up} \over m} {\delta t}^2
$$

- Since the mass is constant we could limit the work to just the accelerations:

$$
x_{t+1} = x_t + v_t \delta t + {a_{forced} - g \over 2} {\delta t}^2
$$

- The velocity will be updated similarly:

$$
v_{t+1} = v_t + (a_{forced} - g) \delta t
$$

The fuzzy engine will take as input the two variables of current distance $d \equiv x - target $ and velocity $v$ and give as output the required **forced acceleration** needed for reaching the goal in a fast and smooth way.

## <font color="green"> Fuzzy Logic </font>

### Main concepts:
- Differs from crisp membership of boolean logic
- **Universe of Discourse**: $x \in U$
- **Membership function**: $\mu(x) \in [0,1]$
- **Fuzzy Set**: $A = \{(x,\mu(x))\mid x \in U\}$
- **Interception of fuzzy sets** (*t-norm*): e.g. $t(a,b) = min(a,b)$
- **Union of fuzzy sets** (*s-norm*): e.g. $s(a,b) = max(a,b)$

### Inference engine (FIE)
- In the sentence "*velocity* is *small*":

    - *velocity* is a **linguistic variable** &rarr; fuzzy variable
    - *small* is a **linguistic value** &rarr; fuzzy set
<br></br>
- Uses fuzzy logic to generate an output to a specific input:
    
    1. **Fuzzyfication** of *crisp* input         
    2. **Inference** of the fuzzy output through **rules**
    3. **Defuzzification** of outptut

<center><img src="Engine.jpg" alt="Drawing" style="width: 750px;" />Figure 2: Schematization of Inference Engine </center>

### Mamdani Inference Engine

- Fuzzy rules in the form: "**IF** ($x_1$ is $X_{i}^1$) and ... and ($x_n$ is $X_{j}^n$) **THEN** $y$ is $Y_k$"
- Each rule produces a certain **Strength** (*fire*)
- The strength of all rules are aggregated &rarr; **Fuzzy output**
- The **fuzzy output** is defuzzified through **Center of Mass** (CoM) or **Middle of Maximum** (MoM) methods

## <font color="green"> Quantum Computing</font>

### Basics:
- Uses **Qubits** as fundamental units
- Two levels quantum state in the 2-dim **Hilbert Space**: $|\psi⟩ = \alpha |0⟩ + \beta |1⟩ \quad \alpha, \beta \in \mathbb{C}$
- Probabilty of measuring qubit in state $|0⟩$ is $|\alpha|^2$ (resp. in $|1⟩$ is $|\beta|^2$)
- Normalization condition imposes: $|\alpha|^2 + |\beta|^2 = 1 $
- Multiple qubits can work together in a tensor product of $n$ Hilbert Spaces:
    
    $$
    H = H_{n-1} ⊗ H_{n-2} \; ⊗ \; ... \; ⊗ \; H_0
    $$
    <br></br>
    This way, a quantum register is written as:
    $$
    |\psi⟩ = \sum_{i=0}^{2^n - 1} \alpha_i |i⟩ \quad \alpha_i \in \mathbb{C}
    $$
    with $|i⟩$ as $i$-th basic vector of the total space.

### Basics:
- A state can be modified without measure through **Unitary Operators**:
    
    $$
    U|\psi⟩ = U[\alpha|0⟩ + \beta|1⟩] = \alpha U|0⟩ + \beta U|1⟩
    $$
    <br></br>
    where **U** satisfies:
    $$
    UU^{\dagger} = U^{\dagger}U = I
    $$
    <br></br>
- Phisically, a Quantum Operator is implemented as a **Quantum Gate** in a **Quantum Circuit**
- Some Quantum operator have a classical counterpart (e.g. **NOT** gate) while others arise exclusively from quantum behaviours (e.g. **Hadamard** gate)
- Each operator has a **matrix representation** in the Hilbert space

<center><img src="Quantum_gates.png" alt="Drawing" style="width: 450px;" />Figure 4: Example of quantum operators </center>

### Quantum Fuzzy Rule Based System (QFRBS):
- A **QFRBS** is defined using **Quantum Oracles**
- A **Quantum Oracle** is a black box. It takes an input $|x⟩$ and returns an output $|h(x)⟩$:

    $$
    U_h|x,\overline{0}⟩ = |x,h(x)⟩
    $$
    <br></br>
- In a **QFRBS** both inputs (fuzzy *antecedents*) and outputs (fuzzy *consequent*) are encoded by binary strings
- Suppose you have $n$ input variables: $X_0, \dots , X_{n-1}$, each defined by $m_j$ linguistic terms: $X_j = \{T_0^j,\dots, T_{m_j - 1}^j\}$
- The output variable $Y$ also has $m_Y$ linguistic terms $Y = \{T_0^Y,\dots, T_{m_Y - 1}^Y\}$

- Each input linguistic term is encoded in a binary string that represents the respective label: 
    - for example: $T_2^j$ will be encoded as $10$
    - the number of bits used for the encoding is the minimum amount needed to encode all linguistic terms of a specific (*j-th*) input variable
    <br></br>
- The output variable uses *one-hot* encoding:
    - for example: $T_3^Y$ is encoded as $001$
    - this time, the number of bit used is equal to the number of linguistic terms $m_y$
<br></br>
- Finally, a fuzzy rule is fed to the oracle $f$ using theese representations: 
    - as an example, the rule: "**IF** ($X_0$ is $T_2^0$) and ($X_1$ is $T_1^1$) **THEN** ($Y$ is $T_3^Y$)" becomes:
    
        $$
        f((10,01)) = 001
        $$

# Quantum Fuzzy Inference Engine (QFIE):

- The only difference with a classical engine is in the **Fuzzy Rules Preprocessing** and the **Inference Operator**
<br></br>
- **Simplest case**:

    - number of linguistic terms of each antecedent is a power of two
    - fuzzy sets are normalized (i.e. no *dummy set* needed)
    - there is a rule for any combination of input linguistic term
<br></br>
- The general workflow of the engine is the following:

**STEP 1**:
<br></br>
Each input is *fuzzified* $\longleftrightarrow$ $\alpha_i^j \equiv $ degree of membership of input $j$ to linguistic term $i$

**STEP 2**:
<br></br>
The initialized state on which the inference wil be computed is created as the superposition:
$$
|\psi_0,\psi_1,\dots,\psi_{n-1}⟩ = \sum_{j=0}^{m_0-1} \sum_{k=0}^{m_1-1} \cdots \sum_{l=0}^{m_{n-1}-1} \sqrt{\alpha_j^0 \alpha_k^1 \cdots \alpha_l^{n-1}} |b_j^0, b_k^1,\dots,b_l^{n-1}⟩
$$
with $b_i^j \longleftrightarrow $ binary encoding of the *i-th* linguistic term of the *j-th* fuzzy variable

**STEP 3**:
<br></br>
For each single rule, the **Oracle** $O_f$ (unitary matrix) acts on the state as follows (supposing $\overline{i},\overline{j},\dots,\overline{k}$ correspond to the antecedent present in such rule):
        
$$
|\psi'_Y⟩ = O_f |\psi_0,\psi_1,\dots,\psi_{n-1}⟩ |\overline{0}⟩
$$
        
$$
|\psi'_Y⟩ = \sum_{i,j,\dots,k \neq \overline{i},\overline{j},\dots,\overline{k}} \sqrt{\alpha_i^0 \alpha_j^1 \cdots \alpha_k^{n-1}}|b_i^0, b_j^1,\dots,b_k^{n-1}⟩|\overline{0}⟩ + \sqrt{\alpha_\overline{i}^0 \alpha_\overline{j}^1 \cdots \alpha_\overline{k}^{n-1}} |b_\overline{i}^0, b_\overline{j}^1,\dots,b_\overline{k}^{n-1}⟩ |c_\overline{l}⟩
$$
with $c_j \longleftrightarrow $ *one-hot* encoding of the *j-th* output linguistic term

**STEP 4**:
<br></br>
All rules can be evaluated simultainously, thus giving as output:
    
$$ 
|\psi'_Y⟩ = O_f |\psi_0,\psi_1,\dots,\psi_{n-1}⟩ |\overline{0}⟩ 
$$
    
$$
|\psi'_Y⟩ = \sum_{i=1}^{m_Y} \sum_{a \in A_S^i} \sqrt{F_a}|a⟩|c_i⟩ 
$$
with $a$ generic antecedent and $A_S^i$ set of antecedents which have consequent linguistic term $T_i^Y$

**STEP 5**:
<br></br>
The probability $P_{c_i}$ of measuring the output state as $|c_i⟩$ is calculated as:
$$
P_{c_i} = \sum_{a \in A_S^i} F_a
$$
(Note that, by nature of quantum measurements, these $P_{c_i}$ cannot be directly computed; however, it is possibile to approximate them by preparing a number of times the state and evaluating the measurement each time, thus reconstructing a discrete probability distribution)

**STEP 6**:
<br></br>
All $P_{c_i}$ are fed to the *aggregator* to infer the **Fuzzy Output**, which, in the end, will be *defuzzyfied* with a decided method

## Fuzzy Sets

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
import QFIE.FuzzyEngines as FE
```

```python
# Definition of universes of discourse
distance = np.linspace(-100, 100, 200)
velocity = np.linspace(-10, 10, 200)
forced_acc = np.linspace(0, 20, 200)

# Membership functions for distance
d_n = fuzz.trapmf(distance, [-100, -100, -10, 0])
d_z = fuzz.trimf(distance, [-10, 0, 10])
d_p = fuzz.trapmf(distance, [0,  10, 100, 100])

# Membership functions for velocity
v_n = fuzz.trapmf(velocity, [-10, -10, -4, 0])
v_z = fuzz.trimf(velocity, [-4, 0, 4])
v_p = fuzz.trapmf(velocity, [0, 4, 10, 10])

# Membership function for forced acceleration
fa_z = fuzz.trapmf(forced_acc, [0, 0, 2.5, 5])
fa_s = fuzz.trapmf(forced_acc, [2.5, 5, 7.5, 10])
fa_g = fuzz.trimf(forced_acc, [7.5, 10, 12.5])
fa_m = fuzz.trapmf(forced_acc, [10, 12.5, 15, 17.5])
fa_h = fuzz.trapmf(forced_acc, [15, 17.5, 20, 20])
```

```python
# Visualize these universes and membership functions
fig, (ax0, ax1, ax2) = plt.subplots(nrows=3, figsize=(10, 12))

# Distance
ax0.plot(distance, d_n, 'b', linewidth=1.5, label='negative')
ax0.plot(distance, d_z, 'g', linewidth=1.5, label='zero')
ax0.plot(distance, d_p, 'r', linewidth=1.5, label='positive')
ax0.set_title('Distance')
ax0.legend()

# Velocity
ax1.plot(velocity, v_n, 'b', linewidth=1.5, label='negative')
ax1.plot(velocity, v_z, 'g', linewidth=1.5, label='zero')
ax1.plot(velocity, v_p, 'r', linewidth=1.5, label='positive')
ax1.set_title('Velocity')
ax1.legend()

# Forced acceleration
ax2.plot(forced_acc, fa_z, 'b', linewidth=1.5, label='zero')
ax2.plot(forced_acc, fa_s, 'g', linewidth=1.5, label='small')
ax2.plot(forced_acc, fa_g, 'black', linewidth=1.5, label='g')
ax2.plot(forced_acc, fa_m, 'r', linewidth=1.5, label='medium')
ax2.plot(forced_acc, fa_h, 'y', linewidth=1.5, label='high')
ax2.set_title('Forced acceleration')
ax2.legend()

# Turn off top/right axes
for ax in (ax0, ax1, ax2):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.show()
```

## Classical Engine

```python
# Initiates the antecedent and consequent variables
dist = ctrl.Antecedent(distance, 'distance')
vel = ctrl.Antecedent(velocity , 'velocity')
f_acc = ctrl.Consequent(forced_acc, 'forced_acc')
```

```python
# Sets the memebership functions
dist['n'] = d_n
dist['z'] = d_z
dist['p'] = d_p

vel['n'] = v_n
vel['z'] = v_z
vel['p'] = v_p

f_acc['zero'] = fa_z
f_acc['small'] = fa_s
f_acc['g'] = fa_g
f_acc['medium'] = fa_m
f_acc['high'] = fa_h
```

```python
# Sets the rules for the engine
rule1 = ctrl.Rule(dist['n'] & vel['n'], f_acc['high'])    
rule2 = ctrl.Rule(dist['n'] & vel['z'], f_acc['medium'])
rule3 = ctrl.Rule(dist['n'] & vel['p'], f_acc['g'])
rule4 = ctrl.Rule(dist['z'] & vel['n'], f_acc['medium'])
rule5 = ctrl.Rule(dist['z'] & vel['z'], f_acc['g'])
rule6 = ctrl.Rule(dist['z'] & vel['p'], f_acc['small']) 
rule7 = ctrl.Rule(dist['p'] & vel['n'], f_acc['g'])
rule8 = ctrl.Rule(dist['p'] & vel['z'], f_acc['small']) 
rule9 = ctrl.Rule(dist['p'] & vel['p'], f_acc['zero']) 
```

```python
# Initiates and builds the engine
f_acc_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9])
f_acc_sys = ctrl.ControlSystemSimulation(f_acc_ctrl)
```

## Quantum Engine

```python
# Sets the rules for the Quantum Engine
rules = ['if distance is n and velocity is n then forced_acc is high',
         'if distance is n and velocity is z then forced_acc is medium',
         'if distance is n and velocity is p then forced_acc is g',
         'if distance is z and velocity is n then forced_acc is medium',
         'if distance is z and velocity is z then forced_acc is g',
         'if distance is z and velocity is p then forced_acc is small',
         'if distance is p and velocity is n then forced_acc is g',
         'if distance is p and velocity is z then forced_acc is small',
         'if distance is p and velocity is p then forced_acc is zero']
```

```python
# Builds the engine
qfie = FE.QuantumFuzzyEngine(verbose=False)

# Adds fuzzy variables
qfie.input_variable(name='distance', range=distance)
qfie.input_variable(name='velocity', range=velocity)
qfie.output_variable(name='forced_acc', range=forced_acc)
```

```python
# Adds membership functions
qfie.add_input_fuzzysets(var_name='distance', set_names=['n', 'z', 'p'], sets=[d_n, d_z, d_p])
qfie.add_input_fuzzysets(var_name='velocity', set_names=['n', 'z', 'p'], sets=[v_n, v_z, v_p])
qfie.add_output_fuzzysets(var_name='forced_acc', 
                          set_names=['zero', 'small', 'g', 'medium', 'high'], 
                          sets=[fa_z, fa_s, fa_g, fa_m, fa_h])

# Adds rules
qfie.set_rules(rules)
```

## <font color="green"> Simulation and comparison </font>

*Helper* functions used to engineer the simulation:

```python
# Function that calculates the next state based on current state and engine output
def next_state(distance, velocity, forced_acc,  timestep=0.01):
    newdist = distance + velocity*timestep - 0.5*10*timestep*timestep + 0.5*forced_acc*timestep*timestep
    newvel = velocity - 10*timestep + forced_acc*timestep
    return newdist, newvel

# Function that makes sure the state boundaries are never surpassed
def bordercheck(input_pos, input_vel, min_distance=-100, max_distance=100, min_velocity=-10, max_velocity=10):
    if input_pos < min_distance: input_pos = min_distance
    if input_pos > max_distance: input_pos = max_distance
    if input_vel < min_velocity: input_vel = min_velocity
    if input_vel > max_velocity: input_vel = max_velocity
    return input_pos, input_vel
```

```python
# Runs a simulation of control of the elevator
def run_simulation(start_position, start_velocity, target, max_timesteps, engine, quantum=True):
    
    input_dist = start_position - target      # distance at the start of the experiment
    input_vel = start_velocity                # input velocity at the start of the experiment
    positions = [input_dist + target] 
    velocities = [input_vel]
    engine_outputs = []
    number_of_timesteps = max_timesteps     

    for i in range(max_timesteps - 1):
    
        if quantum:
            engine.build_inference_qc({'distance':input_dist, 'velocity':input_vel}, draw_qc=False)
            forced_acc = engine.execute(n_shots=1000, plot_histo=False)[0]
        else:
            engine.input['distance'] = input_dist
            engine.input['velocity'] = input_vel
            engine.compute()
            forced_acc = engine.output['forced_acc']

        input_dist, input_vel = next_state(input_dist, input_vel, forced_acc)    # update of the distance and velocity
        input_dist, input_vel = bordercheck(input_dist, input_vel)               # check if within the borders
        positions.append(input_dist + target)
        velocities.append(input_vel)
        engine_outputs.append(forced_acc)

        if i>500 and np.var(positions[-500:]) < 0.01:     # stop iteration if the position has been stablein last 5 seconds
            number_of_timesteps = i + 2
            break
    
    engine_outputs.append(engine_outputs[-1]) # for dimensional compatibilities
    return positions, velocities, engine_outputs, number_of_timesteps
```

### Simulation 3 (gridmap comparison):

```python
# Grid of inputs to visualize ssurfaces of control
distances_list, velocities_list = [t for t in range(-100, 110, 10)], [s for s in range(-10, 11, 1)]

# Builds the output of the quantum engine
f_quantum = []
for d in distances_list:
    for v in velocities_list:
        qfie.build_inference_qc({'distance':d, 'velocity':v}, draw_qc=False)
        f_quantum.append(qfie.execute(n_shots=1000, plot_histo=False)[0])

# Builds the output of the classical engine
f_classical = []
for d in distances_list:
    for v in velocities_list:
        f_acc_sys.input['distance'] = d
        f_acc_sys.input['velocity'] = v
        f_acc_sys.compute()
        f_classical.append(f_acc_sys.output['forced_acc'])
```

```python
# Visualization of the two surfaces
X, Y, = distances_list, velocities_list
plotx, ploty, = np.meshgrid(np.linspace(np.min(X), np.max(X), 21), np.linspace(np.min(Y), np.max(Y), 21))
Z_q = np.zeros_like(plotx)
Z = np.zeros_like(plotx)
counter = 0
for i in range(len(X)):
    for j in range(len(Y)):
        Z_q[i,j] = f_quantum[counter]
        Z[i,j] = f_classical[counter]
        counter += 1
```

```python
# Figure construction
fig = plt.figure(figsize=plt.figaspect(0.5))

# Figure 1: Quantum Surface
ax = fig.add_subplot(1, 2, 1, projection='3d')
plt.title('Quantum Engine')
ax.plot_surface(plotx, ploty, Z_q, rstride=1, cstride=1, cmap='viridis', linewidth=0.4, antialiased=True)
ax.set_xlabel('Distance')
ax.set_ylabel('Velocity')
ax.zaxis.set_rotate_label(False)  # disable automatic rotation
ax.set_zlabel('Forced_acc', rotation=90)
ax.view_init(azim=30)

# Figure 2: Classical Surface
ax = fig.add_subplot(1, 2, 2, projection='3d')
plt.title('Classical Engine')
ax.plot_surface(plotx, ploty, Z, rstride=1, cstride=1, cmap='viridis', linewidth=0.4, antialiased=True)
ax.set_xlabel('Distance')
ax.set_ylabel('Velocity')
ax.zaxis.set_rotate_label(False)  # disable automatic rotation
ax.set_zlabel('Forced_acc', rotation=90)
ax.view_init(azim=30)

plt.show()
```
